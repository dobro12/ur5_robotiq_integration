```python
from control_msgs.msg import GripperCommandAction
from control_msgs.msg import GripperCommandGoal
from control_msgs.msg import GripperCommand
import actionlib
import rospy
```

```python
def connection_timeout(event):
    rospy.logfatal("Gripper seems not to respond")
    rospy.signal_shutdown("Gripper seems not to respond")
```

```python
rospy.init_node('trajectory_testing_client')
```

```python
client = actionlib.SimpleActionClient(
            '/effort_gripper_controller/gripper_cmd', GripperCommandAction)
```

```python
timeout = rospy.Duration(3)
client.wait_for_server(timeout)
```

```python
goal = GripperCommandGoal()
command = GripperCommand()
command.position = 0.0
command.max_effort = 5.0
goal.command = command
```

```python
goal = GripperCommandGoal()
command = GripperCommand()
command.position = 0.8
command.max_effort = 5.0
goal.command = command
```

```python
watchdog = rospy.Timer(rospy.Duration(5.0), connection_timeout, oneshot=True)
client.send_goal(goal)
result = client.wait_for_result()
watchdog.shutdown()
print(result)
```
